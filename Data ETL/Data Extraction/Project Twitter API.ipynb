```python
#import config.py that lists out twitter developer api keys to secure a connection between twitter and python app
from config import *
import tweepy as tw
import pandas as pd
import csv
```

```python
#Authentification process for twitter (tweepy api)
auth = tw.OAuthHandler(TWITTER_CONSUMER_KEY, TWITTER_CONSUMER_SECRET)
auth.set_access_token(TWITTER_ACCESS_TOKEN, TWITTER_ACCESS_TOKEN_SECRET)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
c = tw.Cursor(api.search_tweets, tweet_mode='extended', q='bitcoin', lang="en").items(10)
```

```python
tweet_dataset = pd.DataFrame()
```

```python
for tweet in tw.Cursor(api.search_tweets, tweet_mode='extended', q='bitcoin', lang="en").items(10):
    appending_dataframe = pd.DataFrame([['bitcoin', tweet.id, tweet.created_at, tweet.user.followers_count, tweet.retweet_count, tweet.user.location, tweet.full_text.encode('utf-8')]], columns=['Brand','Tweet Id', 'Tweet Date', 'Follower Count', 'Retweets','Location','Tweet Text'])
    tweet_dataset = tweet_dataset.append(appending_dataframe)
```

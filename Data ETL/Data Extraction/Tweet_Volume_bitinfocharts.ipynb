```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import re

url = 'https://bitinfocharts.com/comparison/bitcoin-tweets.html#3y'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

scripts = soup.find_all('script')
scripts_str = str(scripts[4])
```

```python
scripts
```

```python
split1 = re.findall('\[new Date(.+)\]',scripts_str)[0].split(']')
split2 = [re.sub('[\(\)]|,\[new Date','',i) for i in split1]
split3 = [i.split(',') for i in split2]
```

```python
for i in split3:
    i[0] = re.sub('\"','',i[0])
```

```python
split4 = [i for i in split3 if len(i) == 2]
```

```python
split5 = [i for i in split4 if re.match('[0-9]{4}/[0-9]{2}/[0-9]{2}', i[0])]
```

```python
df = pd.DataFrame(split5, columns = ['date','num_tweets'])
df
```

```python
df.to_csv('tweet_volumes_bitcoin_2014.csv')
```
